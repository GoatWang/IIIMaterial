被式分類器便是將特定條件底下(X=rain, hot, high, false)，球賽開打的機率$P(p|X)$與球賽沒有開打的機率$P(n|X)$進行比較，根據貝是定理: 
$$P(p|X)=P(X|p)·P(p)/P(X)$$ 
$$P(n|X)=P(X|n)·P(n)/P(X)$$
因為是比較，分母可以忽略，因此請算出$P(X|p)·P(p)$以及$P(X|n)·P(n)$，並進行比較:


```python
# 使用df請計算出:
# 1. 球賽開打的機率: P(Postive)
# 2. 球賽沒有開打的機率: P(Negative)
# 3. 如果球賽開打，是晴天的機率: P(Sunny|Postive)
# 4. 如果球賽沒有開打，是晴天的機率: P(Sunny|Negative)
possibilities = {}
#=============your works starts===============#
possibilities["P(Postive)"] = len(df[df['play tennis'] == "P"]) / len(df)
possibilities["P(Negative)"] = len(df[df['play tennis'] == "N"]) / len(df)
possibilities["P(Sunny|Postive)"] = len(df[df['play tennis'] == "P"][df['Outlook'] == "sunny"]) / len(df[df['play tennis'] == "P"])
possibilities["P(Sunny|Negative)"] = len(df[df['play tennis'] == "N"][df['Outlook'] == "sunny"]) / len(df[df['play tennis'] == "N"])
#==============your works ends================#


pprint(possibilities)
# {'P(Negative)': 0.35714285714285715,
#  'P(Postive)': 0.6428571428571429,
#  'P(Sunny|Negative)': 0.6,
#  'P(Sunny|Postive)': 0.2222222222222222}
```

# 3. logistic regression (羅吉斯回歸)
## (1). 理論
- general regression: $\hat{y} = \alpha x_{1} + \beta x_{2} + \gamma x_{3} + bias$

![regression](./imgs/regression.png)

- logistic regression: $\hat{y} = sigmoid(\alpha x_{1} + \beta x_{2} + \gamma x_{3} + bias)$

![logistic_reg](./imgs/logistic_reg.png)

- $logloss = - (ylog(\hat{y}) + (1-y)log(1-\hat{y}))$

- 如何理解logloss
    1. prob
        $$
        h_{k}(x) =   \left\{
        \begin{array}{ll}
        P(y=1|x)=\hat{y}  & \mbox{if}\;\ y=1 \\
        P(y=0|x)=1-\hat{y}  & \mbox{if}\;\ y=0 \\
        \end{array} \right.
        $$
    2. target
        $$
        h_{k}(x) =   \left\{
        \begin{array}{ll}
        \max \hat{y}  & \mbox{if}\;\ y=1 \\
        \max 1-\hat{y}  & \mbox{if}\;\ y=0 \\
        \end{array} \right.
        $$
        
    3. logloss:
        $$
        logloss = - (ylog(\hat{y}) + (1-y)log(1-\hat{y}))\\
        \min logloss
        $$
    
    4. [詳細推倒過程](https://towardsdatascience.com/understanding-binary-cross-entropy-log-loss-a-visual-explanation-a3ac6025181a)
    
    
- why negative and minimize

![Log_vs_neglog](./imgs/Log_vs_neglog.gif)


```python
from sklearn.metrics import log_loss
import numpy as np
print(log_loss([0, 0, 1], [0.0001, 0.0001, 0.9999]))
print(log_loss([0, 0, 1], [0.9999, 0.9999, 0.0001]))

def logloss(y, yhat):
    return -(y)*np.log(yhat)-(1-y)*np.log(1-yhat)
print(logloss(1, 0.9999))
print(logloss(1, 0.0001))

```

## (2). 我們應該先將哪一個特徵值拿來分類
- 分類之後能提供較多資訊量者
- 如何量化資訊量的多寡? Entropy(熵) 
- 包含{m1, m2, …,mn}的M訊息的Entropy(熵)計算如下
$$Entropy(M) = - \sum\limits_{i=1}^n p(m_{i}) log_{2} p(m_{i})$$

$p(m_i)$指的是$m_i$在M裡面出現的機率
- 舉例來說

![entropy](./imgs/entropy.JPG)

![informationGain](./imgs/informationGain.JPG)

- 因此，Information Gain(資訊增量)可以被定義如下:
$$Inpurity_{original} = -\frac{p}{p+n} log_{2}(\frac{p}{p+n}) - \frac{n}{p+n} log_{2}(\frac{n}{p+n})$$
$$Inpurity_{split\_by\_feature} = \sum\limits_{i=1}^v -\frac{p_{i} + n_{i}}{p+n} I(p_{i}, n_{i})$$
$$Information Gain = Inpurity_{original} -Inpurity_{split\_by\_feature}$$


```python
from sklearn.metrics import log_loss
import numpy as np
print(log_loss([0, 0, 1], [0.0001, 0.0001, 0.9999]))
print(log_loss([0, 0, 1], [0.9999, 0.9999, 0.0001]))

def logloss(y, yhat):
    return -(y)*np.log(yhat)-(1-y)*np.log(1-yhat)
print(logloss(1, 0.9999))
print(logloss(1, 0.0001))

```

```python
def entropy(p, n):
    # 請定義出entropy的function，注意若出現0乘上無限大也等於0。
    #=============your works starts===============#
    p_pn = p/(p+n)
    n_pn = n/(p+n)
    lpp = - (p_pn * np.log2(p_pn)) if p_pn != 0 else 0
    lpn = - (n_pn * np.log2(n_pn)) if n_pn != 0 else 0
    entropy = lpp + lpn 
    #==============your works ends================#
    return entropy 

print(entropy(0.5, 0.5))
print(entropy(1, 0))

# print(entropy(0.5, 0.5)*2)
# print(entropy(0.4, 0.5) + entropy(0.6, 0.5))

# 1.0
# 0.0
```

## (2). 參數
- C : float, optional (default=1.0)
    - 錯誤懲罰項
- kernel : string, optional (default='rbf')
    - 決定分隔線的函數: 'linear', 'poly', 'rbf', 'sigmoid', 'precomputed'或是自定義函數
- degree : int, optional (default=3)
    - polynomial(‘poly’)分隔函數的degree，如果使用其他分隔函數將直接被忽略。
- gamma : float, optional (default=’auto’)
    - 'rbf', 'poly' and 'sigmoid'的共變異數. 如果gamma是'auto'則預設為 1/n_features。
